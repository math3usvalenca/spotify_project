```python
import pandas as pd
```

## Pré-processamento dos dados

```python
artists_df = pd.read_csv('artists.csv')
```

```python
artists_df
```

```python
artists_df.isnull().sum()
```

```python
artists_df.info()
```

Filtrando apenas artistas com popularidade maior ou igual a 10.

```python
artists_df = artists_df[artists_df['popularity']>=10]
artists_df
```

Consertando os valores da coluna `genres`

```python
artists_df['genres'] = artists_df['genres'].str.replace('[','').str.replace(']','')
artists_df['genres'] = artists_df['genres'].apply(lambda x: x.replace("'","") )
```

```python
artists_df.head()
```

```python
tracks_df = pd.read_csv('tracks.csv')
```

```python
tracks_df.head()
```

Consertando os valores das colunas `id_artists` e `artists`

```python
tracks_df['id_artists'] = tracks_df['id_artists'].str.replace('[','').str.replace(']','')
tracks_df['id_artists'] = tracks_df['id_artists'].apply(lambda x: x.replace("'","") )
tracks_df['artists'] = tracks_df['artists'].str.replace('[','').str.replace(']','')
tracks_df['artists'] = tracks_df['artists'].apply(lambda x: x.replace("'","") )
```

```python
tracks_df.head()
```

```python
tracks_df.info()
```

Criando uma coluna para o ano do lançamento de uma música

```python
tracks_df['release_date'] = pd.to_datetime(tracks_df['release_date']).dt.year
```

```python
tracks_df.rename(columns={'release_date':'release_year'}, inplace=True)
```

Criando uma coluna para a duração das músicas em minutos

```python
tracks_df['duration_ms']
```

```python
tracks_df['duration_min'] = tracks_df["duration_ms"].apply(lambda x: round(x/60000))
tracks_df.drop("duration_ms", inplace=True, axis=1)
```

```python
tracks_df.drop(['explicit','key','time_signature'],inplace=True, axis=1) 
```

```python
tracks_df.head()
```

Substituindo os nomes de músicas ausentes por 'Unnamed'

```python
tracks_df['name'].replace(pd.NA,'Unnamed', inplace=True) 
```

```python
tracks_df.isnull().sum()
```

Filtrando apenas as músicas com popularidade maior ou igual a 10

```python
tracks_df = tracks_df[tracks_df['popularity']>=10]
tracks_df
```

```python
tracks_df.to_csv('tracks_cleaned.csv',index=False)
```

```python
artists_df.to_csv('artists_cleaned.csv', index=False)
```
